**Sistemas expertos y difusos**

```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# Generar valores uniformes
x_universe = np.linspace(0, 100, 1000)

# Definir las funciones difusas para tres estados: "bajo", "medio" y "alto"
bajo = fuzz.trimf(x_universe, [0, 0, 50])
medio = fuzz.trimf(x_universe, [0, 50, 100])
alto = fuzz.trimf(x_universe, [50, 100, 100])

# Graficar las funciones difusas
fig, ax = plt.subplots()
ax.plot(x_universe, bajo, 'b', linewidth=1.5, label='Bajo')
ax.plot(x_universe, medio, 'g', linewidth=1.5, label='Medio')
ax.plot(x_universe, alto, 'r', linewidth=1.5, label='Alto')
ax.set_title('Funciones de pertenencia')
ax.legend()

plt.show()
```
